```python
max = {'Delhi': 40,
                   'Bengaluru': 31,
                   'Chennai': 35,
                   'Mumbai': 29,
                   'Kolkata': 39}
max = pd.Series(max)
max
```

```python
rainfall = {'Delhi': 24.1,
                   'Bengaluru': 36.2,
                   'Chennai': 40.8,
                   'Mumbai': 35.2,
                   'Kolkata': 41.8}
rainfall = pd.Series(rainfall)
rainfall
```

```python
states = pd.DataFrame({'Max': max,
                       'Min': min,
                      'Rainfall':rainfall})
states
```
